# **Pre-Processing Data in Python**

```python
import pandas as pd #Importamos la libreria de Pandas

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
df = pd.read_csv(url, header=None)  #Creamos el dataframe con el csv de url, no tiene columnas titulos.

# creamos una lista con los headers manualmente para luego agregarlos al df
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df.columns=headers #agregamos a la lista 'headers' como headers.
df.head() #Validamos llamando los primeros 5 registros
```

## Dealing with missing values in Python

```python
df.dropna(axis=0, inplace=True) #elimina los registros nulos,"axis=0": elimina filas, "inplace=True": escribe el resultado directamente en el df

#o podemos usar "df.replace(missing_values, new_value)" para remplazar datos por otros
```

## Data Normalization in Python

**1.Simple feature scaling:** Divide cada valor por el valor máximo para esa función. Esto hace que los nuevos valores oscilen entre 0 y 1.

**2.Min-Max**: Toma cada valor, X_old, restado del mínimo valor de esa característica, luego se divide por el rango de esa característica.
Nuevamente, los nuevos valores resultantes oscilan entre 0 y 1.

**3.z-score or standard score** En esta fórmula, para cada valor, resta el Mu, que es el promedio de la característica, y luego lo divide por la desviación estándar (sigma). Los valores resultantes oscilan alrededor de 0 y, por lo general, oscilan entre -3 y +3, pero pueden ser mayores o menores.

E.g.

```python
df["length"].head
```

```python
df["length"] = df["length"]/df["length"].max() #usamos el primer metodo con la columna 'length'

df["length"].head
```

## Binning in Python

```python
import numpy as np

bins = np.linspace(min(df["price"]), max(df["price"]), num=4) #Primero usamos la función numpy "linspace" para devolver la matriz "bins" que contiene
                                                            #4 números igualmente espaciados sobre el intervalo especificado del precio.
group_names = {"Low","Medium","High"} #Creamos una lista “group_names” que contiene los diferentes nombres de bin.

#Usamos la función de pandas "cut" para segmentar y clasificar los valores de datos en contenedores.
df["price-binned"] = pd.cut(df["price"], bins, labels=group_names, include_lowlest=True) 
```

```python
import numpy as np

#Identify missing values
df.replace("?", np.nan, inplace = True) # replace "?" to NaN

#Evaluating for Missing Data
missing_data = df.isnull()

#Calculate the mean value for the "normalized-losses" column 
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)

df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)

df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)

df['num-of-doors'].value_counts()
df['num-of-doors'].value_counts().idxmax()

#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)

# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
#Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```
